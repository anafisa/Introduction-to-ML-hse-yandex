# `Этапы анализа данных`

 ### 1. Какой тип задачи решаем?

* __классификация__  (классифицируем данные по группам, относим каждый объект к определенному классу)
* __регрессия__  (в качестве ответа получаем вещественное число)
* __кластеризация__ (разбиваем множество объектов на группы, при этом перечень групп и их количество определяется алгоритмом)

### 2. Как измерить качество работы?

* __классификация__: доля правильных ответов, точность, полнота, AUC-ROC
* __регрессия__: средняя квадратичная ошибка, средняя абсолютная ошибка
* __кластеризация__: в зависимости от задачи

### 3. На основе каких данных решаем задачу?

* __числовые признаки__: рост, вес
* __категориальные__: цвет, id 
* __текстовые__: описание товара
* __изображения, сигналы__
* __координаты__

### 4. Формирование матрицы объекты-признаки и предобработка данных

* необходимо избавить данные от __шумов, выбросов, пропущенных значений__

### 5. Построение алгоритма

* __линейные модели__
* __композиции деревьев__: градиентный бустинг, случайный лес
* __нейронные сети__: глубокое обучение, много слоёв

### 6. Оценивание качества

* __отложенная выборка__ 
* __кросс-валидация__ или скользящий контроль (важно правильно подобрать количество блоков)
